# Questão 1

Utilize o dataset Iris disponível no scikit-learn.
Divida os dados em treino e teste utilizando divisão estratificada.

**Solução**:

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42, stratify=y
)

print(f"Total: {len(X)} | Treino: {len(X_train)} | Teste: {len(X_test)}")
```

# Questão 2

Treine um modelo utilizando `DecisionTreeClassifier`.

Depois calcule:

- acurácia no treino
- acurácia no teste

**Solução**:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

modelo = DecisionTreeClassifier(random_state=42)
modelo.fit(X_train, y_train)

y_pred_train = modelo.predict(X_train)
acc_treino = accuracy_score(y_train, y_pred_train)

y_pred_test = modelo.predict(X_test)
acc_teste = accuracy_score(y_test, y_pred_test)

print(f"Acurácia no Treino: {acc_treino * 100:.2f}%")
print(f"Acurácia no Teste: {acc_teste * 100:.2f}%")
```

# Questão 4

Treine dez árvores com:

- max_depth = 1
- max_depth = 2
- max_depth = 3
...
- max_depth = 9
- max_depth = None

Registre em uma tabela para cada árvore:

- acurácia no treino
- acurácia no teste
- profundidade da árvore
- número de folhas

**Solução**:

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

profundidades = list(range(1, 10)) + [None]
resultados = []

for depth in profundidades:
    model = DecisionTreeClassifier(max_depth=depth, random_state=42)
    model.fit(X_train, y_train)
    
    acc_treino = accuracy_score(y_train, model.predict(X_train))
    acc_teste = accuracy_score(y_test, model.predict(X_test))
    
    resultados.append({
        "max_depth_param": str(depth),
        "profundidade_real": model.get_depth(),
        "n_folhas": model.get_n_leaves(),
        "acuracia_treino": acc_treino,
        "acuracia_teste": acc_teste
    })

df_resultados = pd.DataFrame(resultados)
display(df_resultados)
```

**Em qual profundidade começa o overfitting?**
**Por que a árvore consegue 100% no treino quando max_depth=None?**

O overfitting começa na profundidade 4.

Quando você define max_depth=None, você remove o "freio" da árvore, o algoritmo continua criando divisões até que cada "folha" da árvore contenha apenas dados de uma única classe(folhas puras).

# Questão 5

Treine dois modelos:

- criterion = "gini"
- criterion = "entropy"

Compare:

- profundidade da árvore
- acurácia

**Solução**:

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import pandas as pd

modelo_gini = DecisionTreeClassifier(criterion="gini", random_state=42)
modelo_gini.fit(X_train, y_train)

modelo_entropy = DecisionTreeClassifier(criterion="entropy", random_state=42)
modelo_entropy.fit(X_test, y_test) 
modelo_entropy.fit(X_train, y_train)

comparacao = [
    {
        "Criterio": "Gini",
        "Profundidade": modelo_gini.get_depth(),
        "Acuracia_Treino": accuracy_score(y_train, modelo_gini.predict(X_train)),
        "Acuracia_Teste": accuracy_score(y_test, modelo_gini.predict(X_test))
    },
    {
        "Criterio": "Entropia",
        "Profundidade": modelo_entropy.get_depth(),
        "Acuracia_Treino": accuracy_score(y_train, modelo_entropy.predict(X_train)),
        "Acuracia_Teste": accuracy_score(y_test, modelo_entropy.predict(X_test))
    }
]

df_comp = pd.DataFrame(comparacao)
display(df_comp)
```

Desempenho: O critério Gini foi superior, atingindo 93,3% de acurácia no teste, enquanto a Entropia ficou em 88,8%.

Complexidade: O Gini gerou uma árvore mais simples e eficiente com profundidade 5. A Entropia criou um modelo mais complexo, atingindo profundidade 6.

Treino: Ambos os modelos "decoraram" perfeitamente os dados de treino, alcançando 100% de acurácia.

Conclusão: Para este problema, o Gini é a melhor escolha, pois é mais simples e generaliza melhor para dados novos.

# Questão 6

Escolha um hiperparâmetro e investigue seu impacto.

Sugestões:

- max_depth
- min_samples_split
- min_samples_leaf
- criterion

Mostre resultados e interprete.
- melhor modelo encontrado
- acurácia
- parâmetros

**Solução**:

```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

folhas = [1, 2, 5, 10, 20, 30]
res_questão6 = []

for msl in folhas:
    modelo = DecisionTreeClassifier(min_samples_leaf=msl, random_state=42)
    modelo.fit(X_train, y_train)
    
    acc_treino = accuracy_score(y_train, modelo.predict(X_train))
    acc_teste = accuracy_score(y_test, modelo.predict(X_test))
    
    res_questão6.append({
        "min_samples_leaf": msl,
        "profundidade": modelo.get_depth(),
        "acuracia_treino": acc_treino,
        "acuracia_teste": acc_teste
    })

df_q6 = pd.DataFrame(res_questão6)
display(df_q6)
```

hiperparâmetro min_samples_leaf:

Impacto no Modelo: O aumento do min_samples_leaf de 1 para 30 simplificou a árvore, reduzindo sua profundidade de 5 para 2. Isso demonstra que o parâmetro controla a complexidade ao exigir mais dados para criar novas ramificações.

Acurácia: O melhor desempenho foi com min_samples_leaf=1, atingindo 93,3% de acerto no teste. Valores maiores (de 2 a 30) resultaram em uma queda de performance para 88,8%, indicando que a restrição excessiva prejudicou a capacidade de distinção do modelo.

Conclusão: O modelo ideal para este cenário é o de profundidade 5 com o parâmetro em 1. Embora o treino tenha atingido 100%, essa configuração foi a que melhor generalizou para os dados de teste no dataset Iris.
